### Image Processing Fundamentals

#### Image Representation

* A digital image is represented as 2D array of pixels. For grayscale images:
$$
I(x,y) \in [0,225], \quad x \in [0,W-1], \quad y \in [0,H-1]
$$
where w is width, H is height, and intensity range from 0(black) to 225(white)

#### Color Spaces

* **RGB**: Red, Green, Blue channels (0 - 255 each)
* **Grayscale**: Single intensity channel
* **HSV**: Hue, Saturation, Value
Conversion from RGB to Grayscale:
$$ Gray = 0.299R + 0.587G + 0.114B $$
